# Colentado os dados do FED/FRED (Federal Reserve Economic Data)

## 1 - Importando as ferramentas

```python
import pandas as pd
import pandas_datareader.data as web
from scipy.stats import zscore
import plotly.express as px 
import nbformat
import datetime
import warnings
warnings.filterwarnings('ignore')
#pd.options.display.float_format = '{:.6f}'.format
```

```python
start = datetime.datetime(2000,1,1)
```

# 2 - Coletando dados

```python
# PIB USA
gdp = web.DataReader('GDP', 'fred', start)
```

```python
fig = px.line(gdp,x=gdp.index, y='GDP')
```

```python
fig.show()
```

```python
# Inflação USA
cpi = web.DataReader('CPIAUCSL', 'fred', start)
```

```python
fig = px.bar(data_frame=cpi.pct_change()*100,x=cpi.index, y='CPIAUCSL',labels={'CPIAUCSL':'Inflação (%)'})
```

```python
fig.show()
```

```python
# Títulos públicos 3 meses
t3mo = web.DataReader('DGS3MO', 'fred', start)
```

```python
fig = px.line(t3mo,x=t3mo.index, y='DGS3MO')
```

```python
fig.show()
```

```python
# Desemprego
unrate = web.DataReader('UNRATE', 'fred', start)
```

```python
fig = px.line(unrate,x=unrate.index, y='UNRATE')
```

```python
fig.show()
```

```python
var_unrate = unrate.pct_change()[1:]
```

```python
px.bar(data_frame=var_unrate, x=var_unrate.index, y='UNRATE')
```

```python
px.bar(zscore(var_unrate))
```

```python
px.bar(payroll.pct_change())
```
